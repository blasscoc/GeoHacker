# UK Oil and Gas Authority Petroleum Exploration Well Logs
This notebook will load Joined Digital Well Log (JDWL) files made available by the United Kingdom Oil and Gas Authority (OGA) and perform some basic validation checks on the data.  The well logs are periodically released in support of UK Offshore Petroleum Licensing Rounds.  The digital data can be directly downloaded as a zip file from the [OGA digital data portal](https://itportal.ogauthority.co.uk/web_files/gis/Round30Data/OGA/OGA_JDWL_30thRound.zip).

Import modules:

```python
import lasio
from os import path
import numpy as np
import matplotlib.pylab as plt
from os import path
import glob
```

## Find the LAS data

Create an object with the path of the data folder (remember to hash the backslash if using windows):

```python
FOLDERLOC = 'C:\\Users\\Elliot\\Desktop\\Salty Hackathon\\Data\\OGA_JDWL_30thRound\\'
FOLDER = 'NNS'
DATAPATH = FOLDERLOC+FOLDER
```

Search the downloaded data for all files with the _.las_ extension.  Set the recursive flag to 'True' to include subfolders in the search.

```python
counter = 0
for file in glob.glob(path.join(DATAPATH, '*.las'), recursive=True):
    counter += 1
    print(counter, '\t-\t', file.replace(DATAPATH+'\\', ''))
```

## Examine the log curves

Extract the curve mnemonics from the las files and list them against the file name.

```python
# extract the data
las_names = []
curves = []
for file in glob.glob(path.join(DATAPATH, '*.las')):
    las_name = file.replace(DATAPATH + '\\', '') 
    las_names.append(las_name)
    las = lasio.read(path.join(DATAPATH, las_name))
    data = las.df()
    curves.append(list(data.columns))

file_curves = dict(zip(las_names, curves))
    
# print the lists
for l in range(len(las_names)):
    print(las_names[l], ' ', curves[l])
```

Check to see if any standard curves are missing.  This is based on the curve mnemonics in this particular dataset.  Basic set of logs we are looking for are:
- Gamma Ray (GR)
- Caliper (CALI)
- Spontaneous Potential (SP)
- Deep Resistivity (RDEP)
- Shallow Resistivity (RSHAL)
- Density (DENS)
- Neutron (NEUT)
- Sonic (DTC)

Create a dictionary called `incomplete` containing filenames of files without a complete set of well logs as keys and a list of existing well logs as values.

```python
basic_curve_set = ['GR','CALI','RDEP','RSHAL','DENS','NEUT','DTC']
incomplete = {}
for filename, curve in file_curves.items():
    for mnem in basic_curve_set:
        if mnem not in curve:
            incomplete[filename] = curve

for filename, curves in incomplete.items():
    print(filename, ' - ', curves)
```

## Perform basic data validation

Use Lasio to load a las file contained in the folder

```python
las = lasio.read(path.join(DATAPATH, '02_05-10_DATACO_JWDL_QC.las'))
data = las.df()
```

### Gamma Ray Curve
Plot the Gamma Ray curve:

```python
plt.figure(figsize=(5,20))
plt.plot(data['GR'], data.index, '-r', linewidth=0.1, alpha = 1)
plt.ylim(4500,0)
plt.xlabel('Gamma Ray (API)')
plt.ylabel('Depth (m)')
plt.show()
```

Create a histogram that shows the variation of GR readings from the average (median) reading calculated from the 5 readings above and 5 readings below.  This check seeks to identify erroneous spikes in the data - these will show at the limits of the distribution.

```python
med = data['GR'] - data['GR'].rolling(window=5, center=True).median()
indx = ~np.isnan(med)
plt.hist(med[indx],100)
plt.xlim([-20,20])
plt.show()
```

Plot the original GR data and a 'de-spiked' version...

```python
plt.figure(figsize=(5,20))
plt.plot(data['GR'], data.index,'-r', linewidth=0.1, alpha = 0.5)
plt.plot(data['GR'].rolling(window=5, center=True).median(), data.index, '-b', linewidth=0.3)
plt.ylim(4500,0)
plt.show()
```

Plot a subset of the logs to show more details.

```python
top_depth, base_depth = 1000, 1100
plt.figure(figsize=(50,25))
plt.plot(data['GR'][top_depth:base_depth], '-b', alpha=0.5)
plt.plot(data['GR'][top_depth:base_depth].rolling(window=5, center=True).median(), '-r', linewidth=2)
plt.show()
```

### Neutron Density Crossplot

```python
plt.scatter(data['NEUT'], data['DENS'], c=data['GR'], cmap='summer_r', s=5, marker = 'x')
plt.xlabel('Neutron')
plt.ylabel('Bulk Density')
plt.title('Density - Neutron Cross Plot')
plt.colorbar()
plt.show()
```

A very basic validation of the Neutron and Density curves can be acheived by cross plotting the curves and comparing them with Gamma Ray.  In a water filled clastic sequence, the general expectation would be for shale to be relatively high on a Neutron log, with sand being relatively low.  Conversely, a density log would expect to return relatively high values for sandstone, with lower values for shale.  The Gamma Ray log, in a very basic sense, can be used to indicate clastic lithology with relatively high values for shale and relatively low values for sand.  These rules of thumb do not apply where other lithologies are encountered.
